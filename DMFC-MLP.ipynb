```python
X = CD
Y=  df['MA']
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
np.random.seed(22)
hyper_params_mlp = {'alpha':[1e-4, 1e-6, 1e-5, 1e-7, 1e-8]}           
mlp = MLPRegressor(max_iter=200,  learning_rate_init=1, activation='logistic', 
                   learning_rate='invscaling',beta_1=0.8, beta_2=0.7, 
                   random_state=40)
g_search_mlp = GridSearchCV(mlp, hyper_params_mlp, cv=10, n_jobs=-1, )
g_search_mlp.fit(X_Train, Y_Train)
predict_mlp_train = g_search_mlp.predict(X_Train)
predict_mlp_test = g_search_mlp.predict(X_Test)
mse_mlp_test = mean_squared_error(Y_Test,predict_mlp_test)
mse_mlp_train = mean_squared_error(Y_Train,predict_mlp_train)
print(mse_mlp_train)
print(mse_mlp_test)
r2_mlp_train = r2_score(Y_Train,predict_mlp_train)
r2_mlp_test = r2_score(Y_Test,predict_mlp_test)
print(r2_mlp_train)
print(r2_mlp_test)
print(g_search_mlp.best_estimator_)
```
